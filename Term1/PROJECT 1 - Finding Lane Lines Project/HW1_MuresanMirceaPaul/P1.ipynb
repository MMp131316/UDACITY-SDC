## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

x1_prev_right = 0
x2_prev_right = 0
x1_prev_left = 0
x2_prev_left = 0

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def getLaneAngle(line):
    """
    Gets the angle in degrees of the line given as parameter
    """
    angle = 0.0 
    for x1,y1,x2,y2 in line:
        angle = math.atan2(y2-y1, x2-x1) * 180 / math.pi 
    if angle < 0:
        angle = 180 + angle
    
    return angle

def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    resultImage = img
    y_min = image.shape[0] 
    y_max = image.shape[0]
    weight = 0.8
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            if y1 < y_min:
                y_min = y1
                
    #left and right lists for holding 
    left_slope=[] #slope
    left_x=[] #x coordinates
    left_y=[] #y coordinates
    left_top_y = image.shape[0]
    
    right_slope=[]
    right_x=[]
    right_y=[]
    right_top_y=image.shape[0]
    #initialize the thresholds for the angles belonging to the 
    #left and right lanes
    angleRightLow = 30.0
    angleRightHigh = 80.0
    angleLeftLow = 100
    angleLeftHigh = 160
    
    x1_right = 0
    x2_right = 0
    x1_left = 0
    x2_left = 0
    #splitting the lanes into left and right lanes
    for line in lines:
        angle = getLaneAngle(line)
        for x1,y1,x2,y2 in line:
            m = (y2-y1) / (x2-x1) 
            l = math.sqrt(math.pow((x2-x1), 2) + math.pow((y2-y1), 2))
            
            if (m < 0 and angle > angleLeftLow and angle < angleLeftHigh): #left
                left_slope.append(m)
                left_x.append(x1)
                left_x.append(x2)
                left_y.append(y1)
                left_y.append(y2)
                
            if (m > 0 and angle > angleRightLow and angle < angleRightHigh): # right
                right_slope.append(m)
                right_x.append(x1)
                right_x.append(x2)
                right_y.append(y1)
                right_y.append(y2)
    
    #we find the average x and y coordinates of the lanes as well as the slope
    #we use 2 auxiliar variables to hold the previous position of the lane
    #this is useful in case there is a very high variation between the lanes
    #detected in the current frame and the previous found lanes
    global x1_prev_left
    global x2_prev_left 
    if (len(left_slope) > 0):
        avg_left_slope = sum(left_slope) / float(len(left_slope))
        avg_left_x = sum(left_x) / float(len(left_x))
        avg_left_y = sum(left_y) / float(len(left_y))

        b_left = avg_left_y - (avg_left_slope * avg_left_x) 
        x1_left = (y_min - b_left) / avg_left_slope 
        x2_left = (y_max - b_left) / avg_left_slope
       #set the previous variable in 
        if(x1_prev_left == 0):
            x1_prev_left = x1_left 
            
        
        if(x2_prev_left == 0):
            x2_prev_left = x2_left 
       
        #we set the current x value as a weighted function 
        #that depents on the current value of x and the previous value
        x1_left = weight * x1_left + (1 - weight) * x1_prev_left
        x2_left = weight * x2_left + (1 - weight) * x2_prev_left
    
    #verify if the left lane is identified ok
    #in case the it is not store the previous value
    if (np.isnan(x1_left)): x1_left = x1_prev_left
        
    if (np.isnan(x2_left)): x2_left = x2_prev_left
    #verify if the difference between current and
    #previous lanes is not very big. From 2 consecutive frames
    #the variation should not be very big
    if(abs(x1_prev_left - x1_left) > 60):
        x1_left = x1_prev_left
    else:
        x1_prev_left = x1_left 
    if(abs(x2_prev_left - x2_left) > 60):
        x2_left = x2_prev_left
    else:
        x2_prev_left = x2_left 
          
    #do the exact same things as we did for the left lane to the right lane
    global x1_prev_right
    global x2_prev_right
    if (len(right_slope) > 0): 
        avg_right_slope = sum(right_slope)/float(len(right_slope))
        
        avg_right_x = sum(right_x) / float(len(right_x))
        avg_right_y = sum(right_y) / float(len(right_y))
        
        b_right = avg_right_y - (avg_right_slope * avg_right_x) 
        x1_right = (y_min - b_right) / avg_right_slope 
        x2_right = (y_max - b_right) / avg_right_slope 
        
        
        if(x1_prev_right == 0):
            x1_prev_right = x1_right 
         
        if(x2_prev_right == 0):
            x2_prev_right = x2_right 
            
        x1_right = weight * x1_right + (1 - weight) * x1_prev_right
        x2_right = weight * x2_right + (1 - weight) * x2_prev_right 
           
    if (np.isnan(x1_right)): 
        x1_right = x1_prev_right
            
    if (np.isnan(x2_right)): 
        x2_right = x2_prev_right
        
    if(abs(x2_prev_right - x2_right) > 60):
        x2_right = x2_prev_right
    else:
        x1_prev_right = x1_right 
    if(abs(x1_prev_right - x1_right) > 60):
        x1_right = x1_prev_right
    else:
        x2_prev_right = x2_right 
    
    #draw the results onto the image
    poly = np.array( [[[x1_right, y_min],[x2_right,y_max],[x2_left,y_max],[x1_left,y_min]]], dtype=np.int32 )
    cv2.fillPoly(resultImage, poly, color=(0,255,0)) 

    cv2.line(resultImage, (int(x1_right), y_min), (int(x2_right), y_max), color, thickness)
    cv2.line(resultImage, (int(x1_left), y_min), (int(x2_left), y_max), color, thickness)
    
    return resultImage
                
        

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````
